### Ley del semicirculo de Wigner.

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg as LA
import seaborn as sns
```

```python
### Theoretical
def f(x):
    return 1/np.pi * np.sqrt(2-x**2) 
```

```python
### Numerical: GOE, GUE, GSE
n = 8
m = 10000
eig_GOE = []
k = 0
s = np.zeros([(n-1)*m])
for i in range(m):
    H_GOE = np.sqrt(1/n)*np.random.normal(size=(n, n))
    Hsy_GOE = (H_GOE + np.transpose(H_GOE))/2.
    w_GOE,v_GOE = LA.eigh(Hsy_GOE)
    eig_GOE.append(w_GOE)
GOE = np.sort(np.array(eig_GOE).flatten())
sns.histplot(data=GOE,stat="density", element="step", fill=False,kde=True, label='GOE')
eps = 0.0001
t = np.arange(-np.sqrt(2)+eps, np.sqrt(2)-eps, 0.001)
plt.plot(t, f(t), 'k',linewidth=2,label=r'Analitico')
plt.title('Dimension: (n='+str(n)+',m='+str(m)+')')
plt.xlim([-2,2])
plt.ylim([0,0.6])
plt.legend(loc='upper left',bbox_to_anchor = (1, .975)) 

```
